```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Funktion fr att ladda vderdata frn "vder.csv"
def ladda_vaderdata():
    # Lser in CSV-filen med Pandas
    data = pd.read_csv('vder.csv')
    return data

# Funktion fr att analysera temperaturdata
def analysera_data(data, plats1, plats2):
    data_plats1 = data[data['Plats'] == plats1]
    data_plats2 = data[data['Plats'] == plats2]
    medel_temp_plats1 = np.mean(data_plats1['Temperatur'])
    medel_temp_plats2 = np.mean(data_plats2['Temperatur'])
    return medel_temp_plats1, medel_temp_plats2

# Funktion fr att jmfra nederbrd och vindstyrka mellan tv platser
def analysera_nederbord_vind(data, plats1, plats2):
    data_plats1 = data[data['Plats'] == plats1]
    data_plats2 = data[data['Plats'] == plats2]
    medel_nederbord_plats1 = np.mean(data_plats1['Nederbrd'])
    medel_nederbord_plats2 = np.mean(data_plats2['Nederbrd'])
    medel_vindstyrka_plats1 = np.mean(data_plats1['Vindstyrka'])
    medel_vindstyrka_plats2 = np.mean(data_plats2['Vindstyrka'])
    return (medel_nederbord_plats1, medel_nederbord_plats2, medel_vindstyrka_plats1, medel_vindstyrka_plats2)

# Uppdaterad visualiseringsfunktion
def visualisera_data(medel_temp_plats1, medel_temp_plats2, medel_nederbord_plats1, medel_nederbord_plats2, medel_vindstyrka_plats1, medel_vindstyrka_plats2, plats1, plats2):
    # Skapar flera barplots fr att jmfra olika parametrar
    fig, ax = plt.subplots(3, 1, figsize=(8, 12))

    # Temperatur
    ax[0].bar([plats1, plats2], [medel_temp_plats1, medel_temp_plats2], color=['blue', 'lightblue'])
    ax[0].set_ylabel('Medeltemperatur')
    ax[0].set_title('Temperaturjmfrelse')

    # Nederbrd
    ax[1].bar([plats1, plats2], [medel_nederbord_plats1, medel_nederbord_plats2], color=['green', 'lightgreen'])
    ax[1].set_ylabel('Medelnederbrd')
    ax[1].set_title('Nederbrdsjmfrelse')

    # Vindstyrka
    ax[2].bar([plats1, plats2], [medel_vindstyrka_plats1, medel_vindstyrka_plats2], color=['red', 'pink'])
    ax[2].set_ylabel('Medelvindstyrka')
    ax[2].set_title('Vindstyrkejmfrelse')

    plt.tight_layout()
    plt.show()

# Huvudfunktionen fr applikationen
def huvudfunktion():
    plats1 = input("Ange frsta platsen fr jmfrelse: ")
    plats2 = input("Ange andra platsen fr jmfrelse: ")

    data = ladda_vaderdata()
    medel_temp_plats1, medel_temp_plats2 = analysera_data(data, plats1, plats2)
    (medel_nederbrd_plats1, medel_nederbrd_plats2, medel_vindstyrka_plats1, medel_vindstyrka_plats2) = analysera_nederbord_vind(data, plats1, plats2)
    
    visualisera_data(medel_temp_plats1, medel_temp_plats2, medel_nederbrd_plats1, medel_nederbrd_plats2, medel_vindstyrka_plats1, medel_vindstyrka_plats2, plats1, plats2)

# Kr huvudfunktionen
if __name__ == "__main__":
    huvudfunktion()
```
